```python
from tavily import TavilyClient
import os
tavily = TavilyClient(api_key=os.environ['TAVILY_API_KEY'])
try:
    result = tavily.search("test", search_depth="basic")
    print("Tavily API 连接成功")
except Exception as e:
    print("Tavily API 错误:", e)
```

```python
from openai import OpenAI
import os
client = OpenAI(
    api_key=os.environ['AIHUBMIX_API_KEY'],
    base_url=os.environ['AIHUBMIX_BASE_URL']
)
try:
    response = client.chat.completions.create(
        model=os.environ['AIHUBMIX_MODEL_ID_gml_47'],
        messages=[{"role": "user", "content": "Hello"}],
        max_tokens=10
    )
    print("LLM API 连接成功:", response.choices[0].message.content)
except Exception as e:
    print("LLM API 错误:", e)
```
